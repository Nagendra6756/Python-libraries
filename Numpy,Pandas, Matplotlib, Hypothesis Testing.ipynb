```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import sqrt,arange
from scipy import stats
%matplotlib inline
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
student=pd.read_csv('student.csv')
student.head()
```

```python
student.rename(columns={'sex':'gender'},inplace=True)
student['alcohol_index']=(5*student['Dalc']+2*student['Walc'])/7
#Alcohol consumption level
student['ac1']=student['alcohol_index']<=2
student['ac1']=student['ac1'].map({True:'Low',False:'High'})
```

Confidence interval

```python
sample_size=student.shape[0]
sample_size
```

Because we have a sample size greater than 30,we can use central limit theorem to calculate C.I.According to this theorem we can calculate a C.I for the mean using normal distributionn


##To get the confidence interval for the mean we need three numbers:
#1.sample mean
#standard error.
#3.confidence interval
#Formula for standard error:
 #       SE=s/sqrt(n)

```python
sample_mean_grade=student['G3'].mean()
sample_mean_grade
```

```python
std_error_grades=student['G3'].std()/sqrt(sample_size)
```

```python
stats.norm.interval(0.95,loc=sample_mean_grade,scale=std_error_grades)
```

Now lets calculate a C.I for the proportion of students with high alcohol consumption level.Again we need 3 numbers
For proportions the standards error is given by 
SE=sqrtp(1-p)/n

```python
student['ac1'].value_counts(normalize=True)
```

```python
high_prop=student['ac1'].value_counts(normalize=True)['High']
std_error_prop=sqrt(high_prop*(1-high_prop)/sample_size)
```

```python
stats.norm.interval(0.96,loc=high_prop,scale=std_error_prop)
```

25%  seems like a good case for proportion of students with high alcohol consumption

probability calculations

1.Assuming the P(High alc)=0.25.In a class of 10 what is the probability of finding 5 students with high alcohol?

```python
stats.binom.pmf(k=5,n=10,p=0.25)
```

# Hypothesis Testing

Are the population variances equal in the two group of students (low vs high) alcohol consumption?

```python
student.groupby('ac1')['G3'].var()
```

```python
grades_low_ac1=student['G3'][student['ac1']=='Low']
grades_high_ac1=student['G3'][student['ac1']=='High']
stats.bartlett(grades_low_ac1,grades_high_ac1)
```

According to the test we cannot reject null hypothisis

Does alcohol consumption affect academic performances

```python
fig,axes=plt.subplots(1,2,figsize=(14,4))
sns.boxplot(x='ac1',y='G3',data=student,ax=axes[0])
sns.boxplot(x='ac1',y='G3',data=student,ax=axes[1])
```

```python
stats.ttest_ind(grades_low_ac1,grades_high_ac1,equal_var=True)
```

since we got low p value we can reject teh null hypothesis of equal means for the two groups at a level 0.05

Conclusion:There is a statistical signinificant differnece between the grades in the two analyzed groups.since the mean for the group with high alcohol consumption is less than the mean of the other group,the results suggest that alcohol consumption has a negative impact oon students academic performance
